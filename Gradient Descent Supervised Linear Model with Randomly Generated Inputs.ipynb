```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
observation =100000
```

```python
# set the input
```

```python
xs=np.random.uniform(low=-100,high=100,size=(observation,1))
zs=np.random.uniform(low=-100,high=100,size=(observation,1))
```

```python
inputs=np.column_stack((xs,zs))
print(inputs.shape)
```

```python
noise=np.random.uniform(low=-1,high=1,size=(observation,1))
```

```python
# set the target, which weight =[5,-12], and bias=[-11]
```

```python
target=5*xs-12*zs-11+noise
```

```python
# set the initial weight and bias
```

```python
weight=np.random.uniform(low=-0.1,high=0.1,size=(2,1))
bias=np.random.uniform(low=-0.1,high=0.1,size=1)
```

```python
# L2 Approach
```

```python
learning_rate=0.0005
```

```python
for i in range (5000):
    output=np.dot(inputs,weight)+bias
    deltas=output-target
    loss=np.sum(deltas**2)/2/observation
    delta_scaled=deltas/observation
    weight=weight-learning_rate*np.dot(inputs.T,delta_scaled)
    bias=bias-learning_rate*np.sum(delta_scaled)
    print(loss)
```

```python
print (weight)
print(bias)
```

```python
# L1 Approach
```

```python
learning_rate=0.0003
```

```python
for i in range (10000):
    output=np.dot(inputs,weight)+bias
    deltas=output-target
    loss=np.sum(np.absolute(deltas))
    delta_scaled=deltas/observation
    weight=weight-learning_rate*np.dot(inputs.T,delta_scaled)
    bias=bias-learning_rate*np.sum(delta_scaled)
    print(loss)
```

```python
print (weight)
print(bias)
```
